```python
X = [0.7, 0.6, 0.8, 0.5, 0.4, 0.9]


W = [0.2, 0.4, 0.3, 0.1, 0.2, 0.5]
```

```python
# Bias (given)
b = -0.3

# Learning Rate
lr = 0.1

# Target Output (must be Pass)
y = 1

print("Step 1: Given Data")
print("Inputs (X):", X)
print("Weights (W):", W)
print("Bias (b):", b)
print("Target (y):", y)
```

```python
z = 0
for i in range(len(X)):
    z = z + (X[i] * W[i])

z = z + b

print("\nStep 2: Net Input")
print("z =", z)
```

```python
if z >= 0:
    y_pred = 1
else:
    y_pred = 0

print("\nStep 3: Prediction")
print("Predicted Output =", y_pred)
```

```python
error = y - y_pred

print("\nStep 4: Error")
print("Error =", error)
```

```python
if error != 0:
    for i in range(len(W)):
        W[i] = W[i] + (lr * error * X[i])

    print("\nStep 5: Updated Weights")
    print("New Weights =", W)
else:
    print("\nNo Weight Update Needed")
```

```python
b = b + (lr * error)

print("\nStep 6: Updated Bias")
print("New Bias =", b)
```

```python
z_new = 0
for i in range(len(X)):
    z_new = z_new + (X[i] * W[i])

z_new = z_new + b

if z_new >= 0:
    y_new = 1
else:
    y_new = 0
```

```python
print("\nFinal Result:")
if y_new == y:
    print("Target Achieved (Pass = 1)")
else:
    print("Target Not Achieved")
```
